```python
import json
import random
import os
from datetime import datetime

class RecipeApp:
    def __init__(self):
        """Initialize the Recipe App with database loading and setup"""
        self.recipes = []
        self.favorites = []
        self.user_recipes = []
        self.load_database()
        self.load_favorites()
        self.load_user_recipes()
    
    def load_database(self):
        """Load recipes from JSON file"""
        try:
            with open("recipes.json", "r") as f:
                self.recipes = json.load(f)
            print(f"âœ… Loaded {len(self.recipes)} recipes from database")
        except FileNotFoundError:
            print("âŒ recipes.json not found. Please run recipes_database_app.py first.")
            self.recipes = []
    
    def load_favorites(self):
        """Load user favorites from file"""
        try:
            with open("favorites.json", "r") as f:
                self.favorites = json.load(f)
        except FileNotFoundError:
            self.favorites = []
    
    def load_user_recipes(self):
        """Load user-added recipes from file"""
        try:
            with open("user_recipes.json", "r") as f:
                self.user_recipes = json.load(f)
        except FileNotFoundError:
            self.user_recipes = []
    
    def save_favorites(self):
        """Save favorites to file"""
        with open("favorites.json", "w") as f:
            json.dump(self.favorites, f, indent=4)
    
    def save_user_recipes(self):
        """Save user recipes to file"""
        with open("user_recipes.json", "w") as f:
            json.dump(self.user_recipes, f, indent=4)
    
    def display_main_menu(self):
        """Display the main navigation menu"""
        print("\n" + "="*50)
        print("ğŸ³ RECIPE APP - Main Menu")
        print("="*50)
        print("1. ğŸ² Random Recipe Generator")
        print("2. ğŸ“š Browse Recipes by Category")
        print("3. ğŸ” Search by Ingredients")
        print("4. â¤ï¸  My Favorites")
        print("5. â• Add New Recipe")
        print("6. ğŸ“ My Added Recipes")
        print("7. ğŸ  Home Screen")
        print("8. âŒ Exit App")
        print("="*50)
    
    def run(self):
        """Main app loop"""
        print("Welcome to the Recipe App! ğŸ‘‹")
        
        while True:
            self.display_main_menu()
            choice = input("\nSelect an option (1-8): ").strip()
            
            if choice == "1":
                self.random_recipe_generator()
            elif choice == "2":
                self.browse_by_category()
            elif choice == "3":
                self.search_by_ingredients()
            elif choice == "4":
                self.show_favorites()
            elif choice == "5":
                self.add_new_recipe()
            elif choice == "6":
                self.manage_user_recipes()
            elif choice == "7":
                self.home_screen()
            elif choice == "8":
                print("ğŸ‘‹ Thanks for using Recipe App! Goodbye!")
                break
            else:
                print("âŒ Invalid option. Please try again.")

# Initialize the app
app = RecipeApp()
print("Recipe App initialized successfully! ğŸ‰")
```

```python
# Add home screen and random recipe generator methods to the RecipeApp class

def home_screen(self):
    """Display home screen with quick access features"""
    print("\n" + "ğŸ " + "="*48)
    print("ğŸ  HOME SCREEN")
    print("ğŸ " + "="*48)
    
    # Show quick stats
    total_recipes = len(self.recipes) + len(self.user_recipes)
    print(f"ğŸ“Š Total Recipes: {total_recipes}")
    print(f"â¤ï¸  Favorites: {len(self.favorites)}")
    print(f"ğŸ“ My Recipes: {len(self.user_recipes)}")
    
    # Show today's random recipe
    print("\nğŸ² Today's Random Recipe:")
    if self.recipes:
        random_recipe = random.choice(self.recipes)
        print(f"ğŸ½ï¸  {random_recipe['name']} ({random_recipe['category']})")
        print(f"â±ï¸  Cook Time: {random_recipe['cook_time']}")
        print(f"ğŸ¥˜ Ingredients: {', '.join(random_recipe['ingredients'][:3])}...")
        
        # Ask if user wants to see full recipe
        see_full = input("\nWould you like to see the full recipe? (y/n): ").lower()
        if see_full == 'y':
            self.display_recipe_details(random_recipe)
    else:
        print("No recipes available.")
    
    input("\nPress Enter to return to main menu...")

def random_recipe_generator(self):
    """Generate random recipes with options"""
    print("\n" + "ğŸ²" + "="*48)
    print("ğŸ² RANDOM RECIPE GENERATOR")
    print("ğŸ²" + "="*48)
    
    while True:
        print("\nOptions:")
        print("1. ğŸ² Get Random Recipe")
        print("2. ğŸ² Get Random Recipe by Category")
        print("3. ğŸ² Get Multiple Random Recipes")
        print("4. ğŸ”™ Back to Main Menu")
        
        choice = input("\nSelect option (1-4): ").strip()
        
        if choice == "1":
            self.get_single_random_recipe()
        elif choice == "2":
            self.get_random_recipe_by_category()
        elif choice == "3":
            self.get_multiple_random_recipes()
        elif choice == "4":
            break
        else:
            print("âŒ Invalid option. Please try again.")

def get_single_random_recipe(self):
    """Get and display a single random recipe"""
    all_recipes = self.recipes + self.user_recipes
    if not all_recipes:
        print("âŒ No recipes available.")
        return
    
    random_recipe = random.choice(all_recipes)
    print(f"\nğŸ² Your Random Recipe:")
    self.display_recipe_details(random_recipe)
    
    # Ask if user wants to add to favorites
    add_fav = input("\nâ¤ï¸  Add to favorites? (y/n): ").lower()
    if add_fav == 'y':
        self.add_to_favorites(random_recipe)

def get_random_recipe_by_category(self):
    """Get random recipe from specific category"""
    categories = list(set([recipe['category'] for recipe in self.recipes]))
    print(f"\nAvailable categories: {', '.join(categories)}")
    
    category = input("Enter category name: ").strip().lower()
    category_recipes = [r for r in self.recipes if r['category'].lower() == category]
    
    if not category_recipes:
        print(f"âŒ No recipes found in '{category}' category.")
        return
    
    random_recipe = random.choice(category_recipes)
    print(f"\nğŸ² Random {category.title()} Recipe:")
    self.display_recipe_details(random_recipe)

def get_multiple_random_recipes(self):
    """Get multiple random recipes"""
    try:
        num = int(input("How many random recipes? (1-10): "))
        if num < 1 or num > 10:
            print("âŒ Please enter a number between 1 and 10.")
            return
    except ValueError:
        print("âŒ Please enter a valid number.")
        return
    
    all_recipes = self.recipes + self.user_recipes
    if len(all_recipes) < num:
        print(f"âŒ Only {len(all_recipes)} recipes available.")
        return
    
    random_recipes = random.sample(all_recipes, num)
    print(f"\nğŸ² Your {num} Random Recipes:")
    
    for i, recipe in enumerate(random_recipes, 1):
        print(f"\n{i}. {recipe['name']} ({recipe['category']})")
        print(f"   â±ï¸  {recipe['cook_time']}")
        print(f"   ğŸ¥˜ {', '.join(recipe['ingredients'][:3])}...")

def display_recipe_details(self, recipe):
    """Display detailed recipe information"""
    print("\n" + "="*60)
    print(f"ğŸ½ï¸  {recipe['name'].upper()}")
    print("="*60)
    print(f"ğŸ“‚ Category: {recipe['category'].title()}")
    print(f"â±ï¸  Cook Time: {recipe['cook_time']}")
    print(f"ğŸ†” Recipe ID: {recipe['id']}")
    
    print(f"\nğŸ¥˜ Ingredients:")
    for i, ingredient in enumerate(recipe['ingredients'], 1):
        print(f"   {i}. {ingredient.title()}")
    
    print(f"\nğŸ“ Instructions:")
    print(f"   {recipe['instructions']}")
    
    # Check if recipe is in favorites
    if recipe in self.favorites:
        print("\nâ¤ï¸  This recipe is in your favorites!")
    
    print("="*60)

# Add these methods to the RecipeApp class
RecipeApp.home_screen = home_screen
RecipeApp.random_recipe_generator = random_recipe_generator
RecipeApp.get_single_random_recipe = get_single_random_recipe
RecipeApp.get_random_recipe_by_category = get_random_recipe_by_category
RecipeApp.get_multiple_random_recipes = get_multiple_random_recipes
RecipeApp.display_recipe_details = display_recipe_details
```

```python
# Add favorites and search functionality

def add_to_favorites(self, recipe):
    """Add recipe to favorites"""
    if recipe not in self.favorites:
        self.favorites.append(recipe)
        self.save_favorites()
        print(f"â¤ï¸  Added '{recipe['name']}' to favorites!")
    else:
        print(f"â¤ï¸  '{recipe['name']}' is already in your favorites!")

def remove_from_favorites(self, recipe):
    """Remove recipe from favorites"""
    if recipe in self.favorites:
        self.favorites.remove(recipe)
        self.save_favorites()
        print(f"ğŸ’” Removed '{recipe['name']}' from favorites!")
    else:
        print(f"âŒ '{recipe['name']}' is not in your favorites!")

def show_favorites(self):
    """Display and manage favorites"""
    print("\n" + "â¤ï¸" + "="*48)
    print("â¤ï¸  MY FAVORITES")
    print("â¤ï¸" + "="*48)
    
    if not self.favorites:
        print("ğŸ’” No favorites yet. Add some recipes to your favorites!")
        return
    
    while True:
        print(f"\nğŸ“‹ You have {len(self.favorites)} favorite recipes:")
        for i, recipe in enumerate(self.favorites, 1):
            print(f"{i}. {recipe['name']} ({recipe['category']}) - {recipe['cook_time']}")
        
        print("\nOptions:")
        print("1. ğŸ‘€ View Recipe Details")
        print("2. ğŸ—‘ï¸  Remove from Favorites")
        print("3. ğŸ”™ Back to Main Menu")
        
        choice = input("\nSelect option (1-3): ").strip()
        
        if choice == "1":
            try:
                recipe_num = int(input("Enter recipe number: ")) - 1
                if 0 <= recipe_num < len(self.favorites):
                    self.display_recipe_details(self.favorites[recipe_num])
                else:
                    print("âŒ Invalid recipe number.")
            except ValueError:
                print("âŒ Please enter a valid number.")
        elif choice == "2":
            try:
                recipe_num = int(input("Enter recipe number to remove: ")) - 1
                if 0 <= recipe_num < len(self.favorites):
                    recipe = self.favorites[recipe_num]
                    confirm = input(f"Remove '{recipe['name']}' from favorites? (y/n): ").lower()
                    if confirm == 'y':
                        self.remove_from_favorites(recipe)
                else:
                    print("âŒ Invalid recipe number.")
            except ValueError:
                print("âŒ Please enter a valid number.")
        elif choice == "3":
            break
        else:
            print("âŒ Invalid option. Please try again.")

def search_by_ingredients(self):
    """Search recipes by ingredients"""
    print("\n" + "ğŸ”" + "="*48)
    print("ğŸ” SEARCH BY INGREDIENTS")
    print("ğŸ”" + "="*48)
    
    while True:
        print("\nSearch Options:")
        print("1. ğŸ” Search by Single Ingredient")
        print("2. ğŸ” Search by Multiple Ingredients")
        print("3. ğŸ” Advanced Search (Must Have/Can Have)")
        print("4. ğŸ”™ Back to Main Menu")
        
        choice = input("\nSelect option (1-4): ").strip()
        
        if choice == "1":
            self.search_single_ingredient()
        elif choice == "2":
            self.search_multiple_ingredients()
        elif choice == "3":
            self.advanced_search()
        elif choice == "4":
            break
        else:
            print("âŒ Invalid option. Please try again.")

def search_single_ingredient(self):
    """Search by single ingredient"""
    ingredient = input("\nEnter ingredient to search for: ").strip().lower()
    if not ingredient:
        print("âŒ Please enter an ingredient.")
        return
    
    all_recipes = self.recipes + self.user_recipes
    matching_recipes = []
    
    for recipe in all_recipes:
        if any(ingredient in ing.lower() for ing in recipe['ingredients']):
            matching_recipes.append(recipe)
    
    if matching_recipes:
        print(f"\nğŸ” Found {len(matching_recipes)} recipes with '{ingredient}':")
        for i, recipe in enumerate(matching_recipes, 1):
            print(f"{i}. {recipe['name']} ({recipe['category']}) - {recipe['cook_time']}")
        
        # Ask if user wants to see details
        see_details = input(f"\nView recipe details? (enter number or 'n' for no): ").strip()
        if see_details.isdigit():
            try:
                recipe_num = int(see_details) - 1
                if 0 <= recipe_num < len(matching_recipes):
                    self.display_recipe_details(matching_recipes[recipe_num])
            except ValueError:
                pass
    else:
        print(f"âŒ No recipes found with ingredient '{ingredient}'.")

def search_multiple_ingredients(self):
    """Search by multiple ingredients"""
    ingredients_input = input("\nEnter ingredients (separated by commas): ").strip()
    if not ingredients_input:
        print("âŒ Please enter ingredients.")
        return
    
    ingredients = [ing.strip().lower() for ing in ingredients_input.split(',')]
    all_recipes = self.recipes + self.user_recipes
    matching_recipes = []
    
    for recipe in all_recipes:
        recipe_ingredients = [ing.lower() for ing in recipe['ingredients']]
        # Count how many search ingredients are found in this recipe
        matches = sum(1 for search_ing in ingredients if any(search_ing in ing for ing in recipe_ingredients))
        if matches > 0:
            matching_recipes.append((recipe, matches))
    
    if matching_recipes:
        # Sort by number of matches (descending)
        matching_recipes.sort(key=lambda x: x[1], reverse=True)
        print(f"\nğŸ” Found {len(matching_recipes)} recipes with your ingredients:")
        for i, (recipe, matches) in enumerate(matching_recipes, 1):
            print(f"{i}. {recipe['name']} ({recipe['category']}) - {matches} ingredient(s) match")
    else:
        print("âŒ No recipes found with those ingredients.")

def advanced_search(self):
    """Advanced search with must-have and can-have ingredients"""
    print("\nğŸ” Advanced Search:")
    must_have = input("Must have ingredients (separated by commas): ").strip()
    can_have = input("Can have ingredients (separated by commas): ").strip()
    
    must_ingredients = [ing.strip().lower() for ing in must_have.split(',')] if must_have else []
    can_ingredients = [ing.strip().lower() for ing in can_have.split(',')] if can_have else []
    
    if not must_ingredients and not can_ingredients:
        print("âŒ Please enter at least some ingredients.")
        return
    
    all_recipes = self.recipes + self.user_recipes
    matching_recipes = []
    
    for recipe in all_recipes:
        recipe_ingredients = [ing.lower() for ing in recipe['ingredients']]
        
        # Check must-have ingredients
        must_match = True
        if must_ingredients:
            must_match = all(any(must_ing in ing for ing in recipe_ingredients) for must_ing in must_ingredients)
        
        # Count can-have ingredients
        can_matches = 0
        if can_ingredients:
            can_matches = sum(1 for can_ing in can_ingredients if any(can_ing in ing for ing in recipe_ingredients))
        
        if must_match and (not can_ingredients or can_matches > 0):
            matching_recipes.append((recipe, can_matches))
    
    if matching_recipes:
        matching_recipes.sort(key=lambda x: x[1], reverse=True)
        print(f"\nğŸ” Found {len(matching_recipes)} recipes:")
        for i, (recipe, matches) in enumerate(matching_recipes, 1):
            print(f"{i}. {recipe['name']} ({recipe['category']}) - {matches} bonus ingredient(s)")
    else:
        print("âŒ No recipes found matching your criteria.")

# Add these methods to the RecipeApp class
RecipeApp.add_to_favorites = add_to_favorites
RecipeApp.remove_from_favorites = remove_from_favorites
RecipeApp.show_favorites = show_favorites
RecipeApp.search_by_ingredients = search_by_ingredients
RecipeApp.search_single_ingredient = search_single_ingredient
RecipeApp.search_multiple_ingredients = search_multiple_ingredients
RecipeApp.advanced_search = advanced_search
```

```python
# Add browse by category and CRUD functionality

def browse_by_category(self):
    """Browse recipes by category"""
    print("\n" + "ğŸ“š" + "="*48)
    print("ğŸ“š BROWSE RECIPES BY CATEGORY")
    print("ğŸ“š" + "="*48)
    
    # Get all categories
    all_recipes = self.recipes + self.user_recipes
    categories = list(set([recipe['category'] for recipe in all_recipes]))
    categories.sort()
    
    print(f"\nAvailable categories:")
    for i, category in enumerate(categories, 1):
        count = len([r for r in all_recipes if r['category'] == category])
        print(f"{i}. {category.title()} ({count} recipes)")
    
    try:
        choice = int(input(f"\nSelect category (1-{len(categories)}): ")) - 1
        if 0 <= choice < len(categories):
            selected_category = categories[choice]
            self.show_category_recipes(selected_category)
        else:
            print("âŒ Invalid category selection.")
    except ValueError:
        print("âŒ Please enter a valid number.")

def show_category_recipes(self, category):
    """Show recipes in a specific category"""
    all_recipes = self.recipes + self.user_recipes
    category_recipes = [r for r in all_recipes if r['category'] == category]
    
    print(f"\nğŸ“š {category.title()} Recipes ({len(category_recipes)} found):")
    for i, recipe in enumerate(category_recipes, 1):
        print(f"{i}. {recipe['name']} - {recipe['cook_time']}")
    
    while True:
        print(f"\nOptions for {category.title()}:")
        print("1. ğŸ‘€ View Recipe Details")
        print("2. â¤ï¸  Add to Favorites")
        print("3. ğŸ”™ Back to Categories")
        
        choice = input("Select option (1-3): ").strip()
        
        if choice == "1":
            try:
                recipe_num = int(input("Enter recipe number: ")) - 1
                if 0 <= recipe_num < len(category_recipes):
                    self.display_recipe_details(category_recipes[recipe_num])
                else:
                    print("âŒ Invalid recipe number.")
            except ValueError:
                print("âŒ Please enter a valid number.")
        elif choice == "2":
            try:
                recipe_num = int(input("Enter recipe number: ")) - 1
                if 0 <= recipe_num < len(category_recipes):
                    self.add_to_favorites(category_recipes[recipe_num])
                else:
                    print("âŒ Invalid recipe number.")
            except ValueError:
                print("âŒ Please enter a valid number.")
        elif choice == "3":
            break
        else:
            print("âŒ Invalid option. Please try again.")

def add_new_recipe(self):
    """Add a new recipe to user recipes"""
    print("\n" + "â•" + "="*48)
    print("â• ADD NEW RECIPE")
    print("â•" + "="*48)
    
    # Get recipe details
    name = input("Recipe name: ").strip()
    if not name:
        print("âŒ Recipe name is required.")
        return
    
    # Check if recipe already exists
    all_recipes = self.recipes + self.user_recipes
    if any(r['name'].lower() == name.lower() for r in all_recipes):
        print(f"âŒ Recipe '{name}' already exists.")
        return
    
    print("\nAvailable categories: breakfast, fast_food, healthy, vegetarian")
    category = input("Category: ").strip().lower()
    if category not in ['breakfast', 'fast_food', 'healthy', 'vegetarian']:
        print("âŒ Invalid category. Using 'healthy' as default.")
        category = 'healthy'
    
    cook_time = input("Cook time (e.g., '15 minutes'): ").strip()
    if not cook_time:
        cook_time = "Unknown"
    
    print("\nEnter ingredients (one per line, press Enter twice when done):")
    ingredients = []
    while True:
        ingredient = input("Ingredient: ").strip()
        if not ingredient:
            break
        ingredients.append(ingredient)
    
    if not ingredients:
        print("âŒ At least one ingredient is required.")
        return
    
    instructions = input("Instructions: ").strip()
    if not instructions:
        print("âŒ Instructions are required.")
        return
    
    # Create new recipe
    new_id = max([r['id'] for r in all_recipes], default=0) + 1
    new_recipe = {
        'id': new_id,
        'category': category,
        'name': name,
        'ingredients': ingredients,
        'cook_time': cook_time,
        'instructions': instructions,
        'created_by': 'user',
        'created_date': datetime.now().strftime('%Y-%m-%d')
    }
    
    # Add to user recipes
    self.user_recipes.append(new_recipe)
    self.save_user_recipes()
    
    print(f"\nâœ… Recipe '{name}' added successfully!")
    print("Would you like to view the recipe? (y/n): ", end="")
    if input().lower() == 'y':
        self.display_recipe_details(new_recipe)

def manage_user_recipes(self):
    """Manage user-added recipes"""
    print("\n" + "ğŸ“" + "="*48)
    print("ğŸ“ MY ADDED RECIPES")
    print("ğŸ“" + "="*48)
    
    if not self.user_recipes:
        print("ğŸ“ No recipes added yet. Add some recipes first!")
        return
    
    while True:
        print(f"\nğŸ“‹ Your {len(self.user_recipes)} recipes:")
        for i, recipe in enumerate(self.user_recipes, 1):
            print(f"{i}. {recipe['name']} ({recipe['category']}) - {recipe['cook_time']}")
        
        print("\nOptions:")
        print("1. ğŸ‘€ View Recipe Details")
        print("2. âœï¸  Edit Recipe")
        print("3. ğŸ—‘ï¸  Delete Recipe")
        print("4. ğŸ”™ Back to Main Menu")
        
        choice = input("\nSelect option (1-4): ").strip()
        
        if choice == "1":
            try:
                recipe_num = int(input("Enter recipe number: ")) - 1
                if 0 <= recipe_num < len(self.user_recipes):
                    self.display_recipe_details(self.user_recipes[recipe_num])
                else:
                    print("âŒ Invalid recipe number.")
            except ValueError:
                print("âŒ Please enter a valid number.")
        elif choice == "2":
            try:
                recipe_num = int(input("Enter recipe number to edit: ")) - 1
                if 0 <= recipe_num < len(self.user_recipes):
                    self.edit_recipe(recipe_num)
                else:
                    print("âŒ Invalid recipe number.")
            except ValueError:
                print("âŒ Please enter a valid number.")
        elif choice == "3":
            try:
                recipe_num = int(input("Enter recipe number to delete: ")) - 1
                if 0 <= recipe_num < len(self.user_recipes):
                    recipe = self.user_recipes[recipe_num]
                    confirm = input(f"Delete '{recipe['name']}'? (y/n): ").lower()
                    if confirm == 'y':
                        self.delete_recipe(recipe_num)
                else:
                    print("âŒ Invalid recipe number.")
            except ValueError:
                print("âŒ Please enter a valid number.")
        elif choice == "4":
            break
        else:
            print("âŒ Invalid option. Please try again.")

def edit_recipe(self, recipe_index):
    """Edit a user recipe"""
    recipe = self.user_recipes[recipe_index]
    print(f"\nâœï¸  Editing: {recipe['name']}")
    
    # Edit name
    new_name = input(f"Name (current: {recipe['name']}): ").strip()
    if new_name:
        recipe['name'] = new_name
    
    # Edit category
    print(f"Current category: {recipe['category']}")
    new_category = input("Category (breakfast/fast_food/healthy/vegetarian): ").strip().lower()
    if new_category in ['breakfast', 'fast_food', 'healthy', 'vegetarian']:
        recipe['category'] = new_category
    
    # Edit cook time
    new_time = input(f"Cook time (current: {recipe['cook_time']}): ").strip()
    if new_time:
        recipe['cook_time'] = new_time
    
    # Edit ingredients
    print(f"Current ingredients: {', '.join(recipe['ingredients'])}")
    edit_ingredients = input("Edit ingredients? (y/n): ").lower()
    if edit_ingredients == 'y':
        print("Enter new ingredients (one per line, press Enter twice when done):")
        new_ingredients = []
        while True:
            ingredient = input("Ingredient: ").strip()
            if not ingredient:
                break
            new_ingredients.append(ingredient)
        if new_ingredients:
            recipe['ingredients'] = new_ingredients
    
    # Edit instructions
    new_instructions = input(f"Instructions (current: {recipe['instructions']}): ").strip()
    if new_instructions:
        recipe['instructions'] = new_instructions
    
    self.save_user_recipes()
    print("âœ… Recipe updated successfully!")

def delete_recipe(self, recipe_index):
    """Delete a user recipe"""
    recipe = self.user_recipes.pop(recipe_index)
    self.save_user_recipes()
    print(f"ğŸ—‘ï¸  Recipe '{recipe['name']}' deleted successfully!")

# Add these methods to the RecipeApp class
RecipeApp.browse_by_category = browse_by_category
RecipeApp.show_category_recipes = show_category_recipes
RecipeApp.add_new_recipe = add_new_recipe
RecipeApp.manage_user_recipes = manage_user_recipes
RecipeApp.edit_recipe = edit_recipe
RecipeApp.delete_recipe = delete_recipe
```

```python
# Enhanced features: Nutrition facts and dietary tags

def add_nutrition_info(self, recipe):
    """Add nutrition information to recipe display"""
    # Simple nutrition estimation based on ingredients
    nutrition = {
        'calories': self.estimate_calories(recipe),
        'protein': self.estimate_protein(recipe),
        'carbs': self.estimate_carbs(recipe),
        'fat': self.estimate_fat(recipe)
    }
    return nutrition

def estimate_calories(self, recipe):
    """Estimate calories based on ingredients"""
    calorie_ingredients = {
        'egg': 70, 'eggs': 70, 'cheese': 100, 'butter': 100, 'oil': 120,
        'flour': 100, 'bread': 80, 'milk': 50, 'chicken': 200, 'beef': 250,
        'rice': 130, 'pasta': 150, 'potato': 80, 'avocado': 160,
        'banana': 90, 'apple': 50, 'yogurt': 60, 'honey': 60
    }
    
    total_calories = 0
    for ingredient in recipe['ingredients']:
        for key, calories in calorie_ingredients.items():
            if key in ingredient.lower():
                total_calories += calories
                break
    
    return max(total_calories, 50)  # Minimum 50 calories

def estimate_protein(self, recipe):
    """Estimate protein content"""
    protein_ingredients = ['egg', 'chicken', 'beef', 'fish', 'cheese', 'yogurt', 'milk']
    protein_count = sum(1 for ing in recipe['ingredients'] 
                       for prot in protein_ingredients if prot in ing.lower())
    return protein_count * 8  # Approximate grams

def estimate_carbs(self, recipe):
    """Estimate carbohydrate content"""
    carb_ingredients = ['bread', 'flour', 'rice', 'pasta', 'potato', 'banana', 'apple']
    carb_count = sum(1 for ing in recipe['ingredients'] 
                    for carb in carb_ingredients if carb in ing.lower())
    return carb_count * 15  # Approximate grams

def estimate_fat(self, recipe):
    """Estimate fat content"""
    fat_ingredients = ['butter', 'oil', 'cheese', 'avocado', 'egg']
    fat_count = sum(1 for ing in recipe['ingredients'] 
                  for fat in fat_ingredients if fat in ing.lower())
    return fat_count * 10  # Approximate grams

def get_dietary_tags(self, recipe):
    """Get dietary tags for recipe"""
    tags = []
    ingredients = [ing.lower() for ing in recipe['ingredients']]
    
    # Check for dietary restrictions
    if not any(meat in ' '.join(ingredients) for meat in ['chicken', 'beef', 'pork', 'fish', 'meat']):
        tags.append('vegetarian')
    
    if not any(dairy in ' '.join(ingredients) for dairy in ['milk', 'cheese', 'butter', 'yogurt']):
        tags.append('dairy-free')
    
    if not any(gluten in ' '.join(ingredients) for gluten in ['flour', 'bread', 'pasta']):
        tags.append('gluten-free')
    
    if any(vegan_ing in ' '.join(ingredients) for vegan_ing in ['tofu', 'beans', 'vegetables']):
        if 'vegetarian' in tags:
            tags.append('vegan')
    
    return tags

def enhanced_display_recipe_details(self, recipe):
    """Enhanced recipe display with nutrition and dietary info"""
    print("\n" + "="*60)
    print(f"ğŸ½ï¸  {recipe['name'].upper()}")
    print("="*60)
    print(f"ğŸ“‚ Category: {recipe['category'].title()}")
    print(f"â±ï¸  Cook Time: {recipe['cook_time']}")
    print(f"ğŸ†” Recipe ID: {recipe['id']}")
    
    # Add nutrition information
    nutrition = self.add_nutrition_info(recipe)
    print(f"\nğŸ“Š Nutrition (per serving):")
    print(f"   ğŸ”¥ Calories: {nutrition['calories']}")
    print(f"   ğŸ’ª Protein: {nutrition['protein']}g")
    print(f"   ğŸ Carbs: {nutrition['carbs']}g")
    print(f"   ğŸ¥‘ Fat: {nutrition['fat']}g")
    
    # Add dietary tags
    tags = self.get_dietary_tags(recipe)
    if tags:
        print(f"\nğŸ·ï¸  Dietary Tags: {', '.join(tags).upper()}")
    
    print(f"\nğŸ¥˜ Ingredients:")
    for i, ingredient in enumerate(recipe['ingredients'], 1):
        print(f"   {i}. {ingredient.title()}")
    
    print(f"\nğŸ“ Instructions:")
    print(f"   {recipe['instructions']}")
    
    # Check if recipe is in favorites
    if recipe in self.favorites:
        print("\nâ¤ï¸  This recipe is in your favorites!")
    
    print("="*60)

# Update the display_recipe_details method
RecipeApp.add_nutrition_info = add_nutrition_info
RecipeApp.estimate_calories = estimate_calories
RecipeApp.estimate_protein = estimate_protein
RecipeApp.estimate_carbs = estimate_carbs
RecipeApp.estimate_fat = estimate_fat
RecipeApp.get_dietary_tags = get_dietary_tags
RecipeApp.enhanced_display_recipe_details = enhanced_display_recipe_details

# Update the original display_recipe_details to use enhanced version
RecipeApp.display_recipe_details = enhanced_display_recipe_details
```
